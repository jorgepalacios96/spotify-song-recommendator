```python
fulldata = pd.read_csv('dfspoty.csv')
```

```python
fulldata.shape
```

```python
#We round the duration to two decimals

fulldata['duration'] = list(map(lambda x: round(x, 2),fulldata['duration']))
```

```python
fulldata.head(2)
```

```python
#We take only audio features to normalize them

x_data = fulldata[['name', 'authors', 'authors2', 'album', 'duration', 'popularity']]
x_audiofeatures = fulldata[['danceability','energy', 'loudness', 'speechness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']]

from sklearn.preprocessing import StandardScaler

transformer = StandardScaler().fit(x_audiofeatures)
x_scaled = transformer.transform(x_audiofeatures)
dfx_scaled = pd.DataFrame(x_scaled , columns=x_audiofeatures.columns)
dfx_scaled
```

## K-means

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=19,
                init="k-means++",
                n_init=20,  # try with 1, 4, 8, 20, 30, 100...
                max_iter=10,
                tol=0,
                random_state=1234)
kmeans.fit(x_scaled)
print(kmeans.inertia_)
```

```python
# Predicting / assigning the clusters:
clusters = kmeans.predict(x_scaled)

# Check the size of the clusters
pd.Series(clusters).value_counts().sort_index()
```

```python
fulldatascaled = pd.concat([x_data, dfx_scaled], axis = 1)
fulldatascaled.head()
```

```python
fulldatascaled["cluster"] = kmeans.predict(x_scaled)
fulldatascaled.head()
```

## Definition of function

```python
# We load the dataframe with the top100 songs
top100 = pd.read_csv('top100.csv')
```

```python
import random 

def spoty(user_choice):
  x = random.choice(top100['title'])
  if user_choice in list(top100['title']):
    return 'The song you chose is in the top 100! If you like ' + user_choice + ' you may also like '+ x
  elif user_choice in list(fulldatascaled['name']):
    ncluster = fulldatascaled[fulldatascaled['name']==user_choice]['cluster'].values[0]
    datacluster = fulldatascaled[fulldatascaled['cluster']==ncluster]
    return 'You may also like the song ' + random.choice(list(datacluster['name'])) + ' from your playlist.'
  else: 
    return 'The song you chose does not belong to your playlist.'
```
